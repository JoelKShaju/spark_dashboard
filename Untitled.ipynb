```python
import pandas as pd
from pandas import DataFrame as DF
```

```python
cowell_data = pd.read_csv('Cowell FY18-19.csv')
cowell_data
```

```python
cowell_data.columns.values[1] = "Usage"
cowell_data.columns
```

```python
cowell_data['bldg_name'] = "Cowell"
cowell_data['co2_emissions'] = cowell_data['Usage']*0.85
cowell_data
```

```python
cowell_data[['date', 'time']] = cowell_data['Date & Time'].str.split(' ', expand=True)
cowell_data
```

```python
cowell_data.drop(columns=['Date & Time'], inplace=True)
cowell_data
```

```python
def preprocess_df(df, name):
    df.columns.values[1] = "Usage"
    df['bldg_name'] = name
    df['co2_emissions'] = df['Usage']*0.85
    df[['date', 'time']] = df['Date & Time'].str.split(' ', expand=True)
    df.drop(columns=['Date & Time'], inplace=True)
    return df
```

```python
combined_df = cowell_data
```

```python
csi = pd.read_csv('CSI FY18-19.csv')
gillson = pd.read_csv('Gillson FY18-19.csv')
gleeson = pd.read_csv('Gleeson FY18-19.csv')
harney = pd.read_csv('Harney FY18-19.csv')
hayes = pd.read_csv('Hayes Healy FY18-19.csv')
uc = pd.read_csv('UC FY18-19.csv')


df_list = [csi, gillson, gleeson, harney, hayes, uc]
df_names = ['CSI', "Gillson", "Gleeson", "Harney", "Hayes", "UC"]

for df, name in zip(df_list, df_names):
    combined_df = combined_df.append(preprocess_df(df, name), ignore_index=True)
combined_df
```

```python
combined_df.columns.values[3] = "full_date"
combined_df
```

```python
combined_df[['month', 'date', 'year']] = combined_df["date"].str.split('/', expand=True)
combined_df
```

```python
combined_df['full_date']
```

```python
def groupby_day_bldg(combined_df):
    return combined_df.groupby(['full_date', 'bldg_name']).sum()
def groupby_day(combined_df):
    return combined_df.groupby(['full_date']).sum()
```

```python
groupby_day_bldg(combined_df)
```

```python
def groupby_month(combined_df):
    return combined_df.groupby(['month', 'year']).sum()

def groupby_month_bldg(combined_df):
    return combined_df.groupby(['month', 'year', 'bldg_name']).sum()
```

```python
groupby_month_bldg(combined_df)
```

```python
def groupby_year(combined_df):
    return combined_df.groupby(['year']).sum()

def groupby_year_bldg(combined_df):
    return combined_df.groupby(['year', 'bldg_name']).sum()
```

```python
groupby_year_bldg(combined_df)
```

```python
def groupby_week(combined_df):
    combined_df['full_date'] = pd.to_datetime(combined_df['full_date']) - pd.to_timedelta(7, unit='d')
    
    return combined_df.groupby([pd.Grouper(key='full_date', freq='W')]).sum()

def groupby_week_bldg(combined_df):
    combined_df['full_date'] = pd.to_datetime(combined_df['full_date']) - pd.to_timedelta(7, unit='d')
    
    return combined_df.groupby([pd.Grouper(key='full_date', freq='W'), pd.Grouper(key='bldg_name')]).sum()
```

```python
groupby_week_bldg(combined_df)
```
